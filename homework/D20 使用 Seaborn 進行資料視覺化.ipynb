```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 

sns.set_style("whitegrid")

x = np.random.RandomState(7).normal(2, 1, 75)
y = 2 + 1.5 * x + np.random.normal(2, 1, 75)

sns.residplot(x, y, lowess=True, color="red")
```
