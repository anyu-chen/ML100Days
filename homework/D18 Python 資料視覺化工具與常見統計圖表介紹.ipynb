```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)
plt.plot(x,y)
plt.title("y = cosx function")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
plt.savefig("filename.png",dpi=300,format="png")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y, X)

plt.xlim(-2.5,2.5)
plt.ylim(-2.5,2.5)

plt.scatter(X, Y ,c = T, alpha = 0.2)

plt.title("Simple Scatter Plot: arctan2")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```
