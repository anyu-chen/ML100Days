```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
from sklearn import preprocessing

#import pingouin as pg
#!pip install researchpy
#import researchpy   
%matplotlib inline

df_train = pd.read_csv("Titanic_train.csv")
data = df_train.copy()
print(df_train.info())
# Step 1 : Survived,Sex離散轉連續型資料
le = preprocessing.LabelEncoder()
df_train['Survived']=le.fit_transform(data['Survived'])
le = preprocessing.LabelEncoder()
df_train['Sex']=le.fit_transform(data['Sex'])
# Step 2 : 計算資料點的倆倆距離
from sklearn.metrics.pairwise import nan_euclidean_distances
display(pd.DataFrame(nan_euclidean_distances(df_train[['Age', 'Sex', 'Survived', 'Fare']])))
# Step3 : 透過 KNN 進行補值
value_neighbors = 3
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = value_neighbors, weights = 'distance')
df_filled = pd.DataFrame(imputer.fit_transform(df_train[['Age', 'Sex', 'Survived', 'Fare']]))
display(pd.DataFrame(df_filled))
###------------------------------------------------------------------------------------------------###
#Q3: 透過數值法計算 Age 和 Fare 是否有相關性?(連續＆連續)
corr,_ = stats.pearsonr(df_filled[0], df_filled[3])
print("Q3 answer: Age 和 Fare的相關係數 = ",corr)
g = sns.regplot(x = "Age", y = "Fare", color = "b" ,data = df_train)
def judgment_corr(corr):
    if corr < .1:
        qual = 'Negligible'
    elif corr < .39:
        qual = 'Small'
    elif corr < .69:
        qual = 'Medium'
    elif corr < .99:
        qual = 'Large'
    else:
        qual = 'Completely match'
    return(qual)
print("           相關性為 ",judgment_corr(corr))
```
